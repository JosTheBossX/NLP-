```python
import tensorflow as tf
import tensorflow.keras.layers as tf1
# Define the input shape
input_shape = (128, 601, 1)

# Define the model
input_layer = tf.keras.layers.Input(shape=input_shape)

# First conv layer with 32 filters, 3x3 kernel size, and ReLU activation
conv1 = tf1.Conv2D(32, (3, 3), activation='relu', padding='same')(input_layer)

bn1 = tf1.BatchNormalization()(conv1)

maxpool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(bn1)

# Maxpool layer with 2x2 pool size
maxpool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2), padding='same')(conv1)

# Second conv layer with 64 filters and 3x3 kernel size
conv2 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(maxpool1)
bn2 = tf1.BatchNormalization()(conv2)
# Second maxpool layer with 2x2 pool size
maxpool2 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2), padding='same')(bn2)

# Third conv layer with 128 filters and 3x3 kernel size
conv3 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(maxpool2)
bn3 = tf1.BatchNormalization()(conv3)
# Dropout layer to prevent overfitting
dropout = tf.keras.layers.Dropout(0.5)(bn3)

# Fourth conv layer with 256 filters and 3x3 kernel size
conv4 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(dropout)
bn4 = tf1.BatchNormalization()(conv4)
# Adaptive average pooling layer
adaptive_avgpool = tf.keras.layers.GlobalAveragePooling2D()(bn4)

# Define the second input shape
input_shape2 = (15,)

# Define the second input layer
input_layer2 = tf.keras.layers.Input(shape=input_shape2)

# Dense layer with 20 units and ReLU activation
dense = tf.keras.layers.Dense(20, activation='relu')(input_layer2)

# Concatenate the outputs from the two input layers
concat = tf.keras.layers.Concatenate()([adaptive_avgpool, dense])

# Define the model with the two input layers and the concatenated output
model = tf.keras.models.Model(inputs=[input_layer, input_layer2], outputs=concat)
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
import numpy as np
import wave
import contextlib
import glob
import librosa
import librosa.display
import soundfile as sf
import pandas as pd
import os
from pydub import AudioSegment
```

```python
fdir = '/content/drive/MyDrive/AIH/COLANBB/audio_data/combined_data/*.wav'
all_vals = []
for fname in glob.glob(fdir):
    trim_dur = 6.4
    curr_dur = 0
    with contextlib.closing(wave.open(fname,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        curr_dur = frames / float(rate)
        
    if curr_dur < trim_dur:
        
        pad_ms = (trim_dur - curr_dur) # milliseconds of silence needed
        pad_hal_ms = ((pad_ms/2)+0.001)*1000
        f_silence = AudioSegment.silent(duration=pad_hal_ms)
        b_silence = AudioSegment.silent(duration=pad_hal_ms)
        audio = AudioSegment.from_wav(fname)

        padded = f_silence + audio + b_silence  # Adding silence after the audio
        padded.export('/content/drive/MyDrive/AIH/COLANBB/audio_data/padded_data/'+fname.split('/')[-1], format='wav')
        fname = '/content/drive/MyDrive/AIH/COLANBB/audio_data/padded_data/'+fname.split('/')[-1]
    
    data, sr = librosa.load(fname, sr=None, mono=False)
    trimmed = librosa.util.fix_length(data, size=int(sr * 6.4))
    val = librosa.feature.melspectrogram(y=trimmed, sr=sr)   
    all_vals.append(val)
```

```python
final_features = np.stack( all_vals, axis=0 )
print(final_features[0])
```
